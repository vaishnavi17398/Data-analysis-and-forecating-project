```python
# date is object, lets convert the data type to date first, then sort the values and  group the results with same values and then use dense rank
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
```

```python
df['Group'] = df['Date'].rank(method='dense').astype(int)
```

```python
df['Clicks'] = df['Event_Key'].apply(lambda x: x if x == 33 else 0)
df['Impressions'] = df['Event_Key'].apply(lambda x: x if x == 11 else 0)
```

```python
# Filter rows based on Event_Key values
df_clicks = df[df['Event_Key'] == 33]
df_impressions = df[df['Event_Key'] == 11]
```

```python
df_click_count=df_clicks.pivot_table(index='Group', columns='Versions', values='Clicks', aggfunc='count', fill_value=0)
df_click_count
```

```python
df_click_percentage = round(df_click_count.divide(df_click_count.sum(axis=1), axis=0)*100,2)
df_click_percentage.astype(str) + '%'
```

```python
for i in df_click_percentage.columns:
    plt.figure(figsize=(8, 6))
    sns.lineplot(data=df_click_percentage, x=df_click_percentage.index, y=df_click_percentage[i], marker='o')

```

```python
# Melt the DataFrame for plotting using seaborn
df_melted1 = df_click_percentage.reset_index()
df_melted=df_melted1.melt(id_vars='Group', var_name='Versions', value_name='Click_Percentage')
df_melted

```

```python
# Plot the line graph using seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(data=df_melted, x='Group', y='Click_Percentage', hue='Versions', marker='o')
```

```python
Inference : The number of clicks for the 'Control' version increases with the ascending group numbers,
    indicating a positive correlation between the group number and the associated date, 
    suggesting a daily increase in the number of clicks.
    
The 'T2' version experiences a rise in clicks until 2019-05-02, remaining constant until 2019-05-03.

Similarly, the 'T3' version witnesses an increase in clicks until 2019-05-02, followed by a decrease until 2019-05-03.

The rate of decline in clicks for the 'T1' version is more pronounced from May 1, 2019, to May 2, 2019, compared to the decline from May 2, 2019, to May 3, 2019.
```
